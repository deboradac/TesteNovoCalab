Explicação if e elseif

```python
# Lista de produtos com seus estoques
estoque = {
    "notebook": 5,
    "mouse": 12,
    "teclado": 7,
    "monitor": 3
}

# Verifica o estoque de cada produto usando for e if
print("Verificando estoque dos produtos:")
for produto, quantidade in estoque.items():
    if quantidade == 0:
        print(f"{produto}: fora de estoque")
    elif quantidade < 5:
        print(f"{produto}: estoque baixo ({quantidade} unidades)")
    else:
        print(f"{produto}: estoque ok ({quantidade} unidades)")

# Simula a venda de produtos usando while
print("\nSimulando vendas:")
vendas = ["mouse", "monitor", "monitor", "monitor", "monitor"]

i = 0
while i < len(vendas):
    produto = vendas[i]
    if produto in estoque and estoque[produto] > 0:
        estoque[produto] -= 1
        print(f"Venda realizada: {produto} (restam {estoque[produto]} unidades)")
    else:
        print(f"Venda não realizada: {produto} está fora de estoque")
    i += 1

# Resultado final do estoque
print("\nEstoque final:")
for produto, quantidade in estoque.items():
    print(f"{produto}: {quantidade} unidades")
```

```python
# Inicializa o estoque com alguns produtos
estoque = {
    "notebook": 5,
    "mouse": 12,
    "teclado": 7,
    "monitor": 3
}

# Verifica o estoque atual
print("Estoque atual:")
for produto, quantidade in estoque.items():
    if quantidade == 0:
        print(f"{produto}: fora de estoque")
    elif quantidade < 5:
        print(f"{produto}: estoque baixo ({quantidade} unidades)")
    else:
        print(f"{produto}: estoque ok ({quantidade} unidades)")

# Simula vendas com entrada do usuário
print("\nDigite os produtos vendidos (digite 'sair' para encerrar):")
while True:
    produto_vendido = input("Produto vendido: ").lower()
    if produto_vendido == "sair":
        break
    if produto_vendido in estoque:
        if estoque[produto_vendido] > 0:
            estoque[produto_vendido] -= 1
            print(f"Venda realizada: {produto_vendido} (restam {estoque[produto_vendido]} unidades)")
        else:
            print(f"{produto_vendido} está fora de estoque")
    else:
        print("Produto não encontrado no estoque")

# Mostra o estoque final
print("\nEstoque final:")
for produto, quantidade in estoque.items():
    print(f"{produto}: {quantidade} unidades")
```
